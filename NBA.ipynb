**WITH SELENIUM WE SCRAPE DATA FROM JAVASCRIPT PAGES AND PUT IT INTO A HTML FILE**
**NOW THAT WE HAVE HTML PAGES WITH DATA THAT WE REQUIRE WE MAKE USE OF BEAUTIFULSOUP4**
**NOW BEAUTIFULSOUP4 WILL PARSE THESE HTML PAGES THAT WE CREATED USING SELENIUM**


**FIRST WE TRY THIS FOR A SINGLE PAGE (1991)**

```python
year = 1991
with open(f"player/{year}.html", encoding="utf-8") as f:
    page=f.read()
    
soup = BeautifulSoup(page, "html.parser")
soup.find("tr", class_= "thead").decompose()
player_table = soup.find(id="per_game_stats")

player_1991 = pd.read_html(str(player_table))[0] 
player_1991["Year"] = year    # Add an 'Year' column to the data

```

```python
player_1991
```

**SUCCESSFULLY DONE FOR A SINGLE PAGE (1991)**
**NOW WE DO THIS FOR ALL THE PAGES (1991-2023)**

```python
df = []
for year in years:
    with open(f"player/{year}.html", encoding="utf-8") as f:
        page=f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find("tr", class_= "thead").decompose()
    player_table = soup.find(id="per_game_stats")
    player = pd.read_html(str(player_table))[0] 
    player["Year"] = year    # Add an 'Year' column to the data
    df.append(player)

# 'df.append(player)' creates a list of multiple dataframes by appending dataframe of
# player statistics of each year (1991-2023)
# So we have a list of dataframes now stored in 'df'
```

**Scraping Team Records which will be later used to predict the MVP**

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

**FIRST TRY FOR A SINGLE YEAR**

```python
year = 1991
url = team_stats_url.format(year)
data = requests.get(url)
with open(f"team/{year}.html", "w+", encoding="utf-8") as f: 
        f.write(data.text)
```

**NOW FOR MULTIPLE YEARS**

```python
for year in years:
    url = team_stats_url.format(year)
    data = requests.get(url)
    with open(f"team/{year}.html", "w+", encoding="utf-8") as f: 
        f.write(data.text)
    
```

**NOW SCRAPING ONLY THE DATA THAT WE WANT USING BS4**

```python
df = []
for year in years:
    with open("team/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find("tr", class_= "thead").decompose()
    team_table = soup.find(id="divs_standings_E")
    team = pd.read_html(str(team_table))[0]
    team["Year"] = year
    team["Team"] = team["Eastern Conference"]
    del team["Eastern Conference"]
    df.append(team)
    
    soup = BeautifulSoup(page, "html.parser")
    soup.find("tr", class_= "thead").decompose()
    team_table = soup.find(id="divs_standings_W")
    team = pd.read_html(str(team_table))[0]
    team["Year"] = year
    team["Team"] = team["Western Conference"]
    del team["Western Conference"]
    df.append(team)
```

```python
teams = pd.concat(df)
```

```python
teams
```

```python
teams.to_csv("teams.csv")
```
