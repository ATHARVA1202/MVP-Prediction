**MOST ML MODELS DON'T RUN WELL WITH NULL VALUES IN ITS DATA.**

**SO NOW WE REMOVE ANY NULL VALUES PRESENT IN THE DATA**

```python
pd.isnull(stats)
```

```python
pd.isnull(stats).sum()
```

**WE  SEE UNDEFINED OR MISSING VALUES IN THE PERCENTAGES(%) FIELDS**

**NOW ANALYSING THOSE FIELDS**

```python
stats[pd.isnull(stats["3P%"])]
```

```python
stats[pd.isnull(stats["3P%"])][["Player", "3PA"]]
```

```python
stats[pd.isnull(stats["FT%"])][["Player", "FTA"]]
```

**WE SEE HERE THAT PLAYERS WITH ANY UNDEFINED OR MISSING VALUES FOR '3 POINT % (3P%)' HAVE 0 '3 POINT ATTEMPTS (3PA)'** 

**ALSO THAT PLAYERS WITH ANY UNDEFINED OR MISSING VALUES FOR 'FT %' HAVE 0 'FT ATTEMPTS'** 

**SO NOW WE REPLACE THESE MISSING VALUES IN '3 POINT % (3P%)' WITH 0**

**EVEN THOUGH THIS IS NOT TECHNICALLY CORRECT BUT STILL A PLAYER WITH NO '3 POINT ATTEMPTS (3PA)' IS NOT POSSIBLE TO WIN THE MVP ANYWAY**

```python
stats = stats.fillna(0)
```

```python
pd.isnull(stats).sum()
```
