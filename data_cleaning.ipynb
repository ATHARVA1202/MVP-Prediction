```python
players = pd.read_csv("players.csv")
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

**WE SEE "*" CHARACTER IN A LOT OF PLACES LIKE IN FRONT OF CHARLES BARKLEY AND LARRY BIRD**

```python
players["Player"] = players["Player"].str.replace("*","",regex=False)
```

```python
players.head(50)
```

**WE SEE MULTIPLE NAMES OF SINGLE PLAYER. WE JUST NEED THE TOTAL WHICH IS THE FIRST NAME IN THE DATAFRAME** 

```python
players.groupby(["Player","Year"]).get_group(("Greg Anderson",1991))
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]   # Finds in row 'Tm' with value 'TOT' 
        row["Tm"] = df.iloc[-1,:]["Tm"]  # Replaces the name 'TOT' with last element which is the latest team of the player for that year
        return row

players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

**Drop levels twice to remove the unwanted two levels**

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players.head(20)
```

**PERFORM AN OUTER MERGE FOR TWO TABLES "players" AND "mvps"**

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])  
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```
